```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://punchng.com/topics/news/'
response = requests.get(url).text
```

```python
soup = BeautifulSoup(response, 'html.parser')
article_container = soup.find('div', class_='latest-news-timeline-section')
article_container.find('article')
```

```python
article_container_temp = article_container.find_next('article')
article_container_temp
```

```python
article_container_temp = article_container.find_all_next('article')
article_container_temp
```
